## Import Model

```python
from sklearn.cluster import KMeans
```

```python
k_model=KMeans(n_clusters=4,random_state=42)
```

```python
k_model.fit(data)
```

```python
yp=k_model.fit_predict(data)
```

```python
data["Cluster"]=yp
```

```python
data
```

```python
data["Cluster"].unique()
```

```python
data["Cluster"].replace([2, 3, 0, 1],["High_Risk","Very_High_Risk","Low_Risk","Medium_Risk"],inplace=True)
```

```python
print("Cluester_Centers:\n\n" ,k_model.cluster_centers_)
```

```python
data
```

```python
cluster_means=data.groupby("Cluster").mean()
print(cluster_means)
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
sns.heatmap(cluster_means,annot=True,cmap="coolwarm")
```
